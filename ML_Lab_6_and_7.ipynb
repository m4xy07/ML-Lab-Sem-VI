**AIM: Implement feature techniques(Standardization, Min-Max Scaling ) and observe their effect on model performance**

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, classification_report
```

```python
import pandas as pd
data = load_wine()
x=pd.DataFrame(data.data,columns=data.feature_names)
y=data.target
#split
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.3,random_state=42
)
```

```python
print("\n----- BEFORE SCALING -----")
print(x_train.head())
```

```python
#KNN without Scaling
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
acc_no = accuracy_score(y_test,knn.predict(x_test))
print("\nAccuracy WITHOUT scaling:",acc_no)
```

```python
#Standard Scaling
scaler_std = StandardScaler()
x_train_std = scaler_std.fit_transform(x_train)
x_test_std = scaler_std.transform(x_test)
x_train_std_df = pd.DataFrame(x_train_std, columns=x_train.columns)
```

```python
print("\n----- AFTER  STANDARD SCALING -----")
print(x_train_std_df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
#Min-Max Scaling (0-1 range)
scaler_minmax = MinMaxScaler()
x_train_mm = scaler_minmax.fit_transform(x_train)
x_test_mm = scaler_minmax.transform(x_test)
x_train_mm_df= pd.DataFrame(x_train_mm, columns=x_train.columns)
print("\n----- AFTER MINMAX SCALING -----")
print(x_train_mm_df.head())
```

```python
knn.fit(x_train_std, y_train)
acc_std = accuracy_score(y_test, knn.predict(x_test_std))
print("\nAccuracy WITH Standard Scaling:", acc_std)
```

```python
scaler_minmax = MinMaxScaler()
x_train_mm = scaler_minmax.fit_transform(x_train)
x_test_mm = scaler_minmax.transform(x_test)
x_train_mm_df = pd.DataFrame(x_train_mm, columns=x_train.columns)
print(x_train_mm_df.head())
```

```python
knn.fit(x_train_mm, y_train)
acc_mm = accuracy_score(y_test, knn.predict(x_test_mm))
print("\nAccuracy WITH MinMaxScaler:", acc_mm)
```

```python
#Confusion Matrix
cm = confusion_matrix(y_test, knn.predict(x_test_mm))
print("Confusion Matrix:")
print(cm)
#Accuracy
acc= accuracy_score(y_test, knn.predict(x_test_mm))
print("\nAccuracy:", acc)
#Precision
prec = precision_score(y_test, knn.predict(x_test_mm), average='macro')
print("\nPrecision:", prec)
#Recall
rec = recall_score(y_test, knn.predict(x_test_mm), average='macro')
print("\nRecall:", rec)
#F1 Score
f1 = f1_score(y_test, knn.predict(x_test_mm), average='macro')
print("\nF1 Score:", f1)
#full classification report
print("\nClassification Report:")
print(classification_report(y_test, knn.predict(x_test_mm)))
```
