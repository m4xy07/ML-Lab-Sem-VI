● Wrapper Method (Recursive Feature Elimination – RFE)

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("/content/housing.csv")
X = df.drop("median_house_value", axis = 1)
y = df["median_house_value"]
print(X.shape)
```

```python
df = X.copy()
```

```python
df['median_house_value'] = y
df_numeric = df.select_dtypes(include = 'number')
target_corr = df_numeric.corr()['median_house_value'].drop('median_house_value')
print("\nCorrelation with Target:")
print(target_corr.sort_values(ascending = False))
selected_features = target_corr[abs(target_corr) > 0.3]
print(selected_features)
```

```python
from sklearn.linear_model import LinearRegression

X_numeric = X.select_dtypes(include='number').fillna(X.select_dtypes(include='number').median())
y_numeric = y

model = LinearRegression()
rfe = RFE(estimator = model, n_features_to_select = 5)
rfe.fit(X_numeric, y_numeric)
print("\nSelected Features")
print(X_numeric.columns[rfe.support_])
```
