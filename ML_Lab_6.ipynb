```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Train KNN on original (unscaled) data
knn.fit(X_train, y_train)

# Make predictions on original test data
y_pred_original = knn.predict(X_test)

# Evaluate performance on original data
accuracy_original = accuracy_score(y_test, y_pred_original)
print(f"Accuracy on original data: {accuracy_original:.4f}")
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
minmax_scaler = MinMaxScaler()

# Apply MinMaxScaler to the training data
X_train_scaled_minmax = minmax_scaler.fit_transform(X_train)

# Apply MinMaxScaler to the test data
X_test_scaled_minmax = minmax_scaler.transform(X_test)

# Display the first 5 rows of the Min-Max scaled training data
print("Min-Max Scaled X_train (first 5 rows):\n", X_train_scaled_minmax[:5])
print("\nMin of X_train_scaled_minmax:\n", X_train_scaled_minmax.min(axis=0))
print("\nMax of X_train_scaled_minmax:\n", X_train_scaled_minmax.max(axis=0))
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the training data
X_train_scaled_standard = scaler.fit_transform(X_train)

# Apply StandardScaler to the test data
X_test_scaled_standard = scaler.transform(X_test)

# Display the first 5 rows of the standardized training data
print("Standardized X_train (first 5 rows):\n", X_train_scaled_standard[:5])
print("\nMean of X_train_scaled_standard:\n", X_train_scaled_standard.mean(axis=0))
print("\nStandard Deviation of X_train_scaled_standard:\n", X_train_scaled_standard.std(axis=0))
```
